Please see [github page](https://github.com/ykatsu111/jupyter-grads-kernel) for installation.

# Usage

This kernel is a simple wrapper of GrADS, so that you can execute any commands of GrADS.  

At the beginning, you need to open ctl file as well as the original GrADS.


```grads
sdfopen anl_surf125.nc
```

Of course, you can execute any GrADS commands such as q XXXX.

```grads
q ctlinfo
```

You can display grads data completely same as original grads command line.  
The display will be shown after 'display' or 'd' command.

```grads
set grads off
set timelab off
set t 2
d PRMSL_GDS0_MSL / 100
```

Since the default background color is black, changing it to white is probably better.

```grads
c
set grads off
set background 1
set t 5
set gxout shaded
d PRMSL_GDS0_MSL / 100
```

If some grads functions or libraries are installed on your system, you can call them.  
For example, cbarn function will draw color bar on the display.

```grads
c
set grads off
set t 10
set gxout shaded
d PRMSL_GDS0_MSL / 100
cbarn
```

If the first line of cell starts with `*%script`, the cell will be interapted as GrADS script.  
In this case, `*%display` is also write at a line if you would like to shown the display after running script.  
The magic command `*%display` works to force to show display.

```grads
*%script
*%display
'reinit'
say 'This is GrADS script!'
'sdfopen anl_surf125.nc'
'set grads off'
'set timelab off'
'set background 1'
'set t 15'
'd PRMSL_GDS0_MSL / 100'
```

# Limitation

* Animation is not supported yet, so that you may not call 'd' or 'display' when some variables are variating. If you do so, probably the last state of display will be shown on you notebook.
